# 6.Cleanse the data to make the year, population, cost, cost per person numerical type - use apply() function

```python
def cleanse_year(year_str):
  n = "*" in year_str
  if n == True:
       year=year_str.replace("*","")
  else:
          year=year_str
  return int(year)

df["Census Year"]=df["Census Year"].apply(cleanse_year)

df
```

```python
def cleanse_pop(pop_str):
  n = "," in pop_str
  if n == True:
       pop=pop_str.replace(",","")
  else:
          pop=pop_str
  return int(pop)

df["Total Population"]=df["Total Population"].apply(cleanse_pop)

df
```

```python
df.iloc[1,2].replace('$','')
```

```python
census_cost = []
avg_cost = []
for i in range(len(df)):
    census_cost.append(df.iloc[i,2].replace('$','').replace(' Billion',''))
    avg_cost.append(df.iloc[i,3].replace(' cents','').replace('$',''))
```

```python
df['Census Cost']=census_cost
df['Average Cost Per Person']=avg_cost
df
```
